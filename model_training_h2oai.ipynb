```python
import h2o
from h2o.automl import H2OAutoML
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
import os
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data_preprocessed.csv')
df.head()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
y = df['Approved_Flag']
x = df.drop(['Approved_Flag'], axis = 1 )
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
h2o.init()
```

```python
x_train = h2o.H2OFrame(df)
x_train
```

```python
x_test = h2o.H2OFrame(x_test)
y_test = 'Approved_Flag'
```

```python
aml = H2OAutoML(max_runtime_secs = 300,
                exclude_algos =['DeepLearning'],
                seed = 1,
                # stopping_metric ='logloss',
                # sort_metric ='logloss',
                balance_classes = False,
)
# train model and record time % time 
aml.train(x = x_test.columns, y = y_test,training_frame=x_train)
```

```python
lb = h2o.automl.get_leaderboard(aml, extra_columns = "ALL")
lb
```

```python
# Get the top model of leaderboard
se = aml.leader
 
# Get the metalearner model of top model
metalearner = h2o.get_model(se.metalearner()['name'])
 
# list baselearner models :
metalearner.varimp()
```
